```python
import numpy as np
np.random.seed(1)
x=np.random.randint(0,50,1000)
print(x)
print("y")
y=x-np.random.randint(0,10,1000)
print(y)
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
x1=np.random.randint(0,50,1000)
y1=100-x1+np.random.normal(0,5,1000)
np.corrcoef(x1,y1)
```

```python
plt.scatter(x1,y1)
plt.show()
```

```python
z=np.random.randint(0,100,1000)
print('z and x')
np.corrcoef(x,z)
plt.scatter(x,z)
plt.show()
print('z and y')
np.corrcoef(y,z)
plt.scatter(y,z)
plt.show()
print('z and x1')
np.corrcoef(x1,z)
plt.scatter(x1,z)
plt.show()
print('z and y1')
np.corrcoef(y1,z)
plt.scatter(y1,z)
plt.show()
```

```python
import pandas as pd
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.randint(0,50,1000)
df['c']=100-df['a']+np.random.randint(0,5,1000)
df['d']=np.random.randint(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
